Precision and Recall: These metrics are commonly used in binary classification problems, where the goal is to detect anomalies (the positive class) among a larger set of non-anomalies (the negative class). Precision measures the proportion of true positives among all predicted positives, while recall measures the proportion of true positives among all actual positives. Precision and recall can be computed as:

Precision = TP / (TP + FP)
Recall = TP / (TP + FN)

where TP is the number of true positives (i.e., anomalies correctly detected), FP is the number of false positives (i.e., non-anomalies incorrectly detected as anomalies), and FN is the number of false negatives (i.e., anomalies that were not detected).

F1 Score: The F1 score is a measure that balances precision and recall, and is defined as the harmonic mean of the two metrics. The F1 score ranges from 0 to 1, with a higher score indicating better performance. The F1 score can be computed as:

F1 Score = 2 * (Precision * Recall) / (Precision + Recall)

DBSCAN works by defining neighborhoods based on the density of points and connecting nearby neighborhoods to form clusters.

Local outliers, also known as point anomalies, are individual data points that are considered unusual within a small neighborhood or cluster of points. They are characterized by their deviation from the surrounding points within a local region. 

Global outliers, also known as contextual anomalies, are data points that are unusual or unexpected within the entire dataset. They are characterized by their deviation from the overall distribution of the data. 

The main difference between local and global outliers is the scope of their impact. Local outliers have a limited impact on the dataset, as they only affect a small neighborhood or cluster of points. On the other hand, global outliers have a broader impact on the dataset, as they affect the overall distribution of the data.
